```python
import pandas as pd
df = pd.read_csv("Students.csv")
print(df)

# Find missing values
print("Missing values:\n", df.isnull().sum(), "\n")

# Fill missing marks with column mean
df.fillna({"Math" : df["Math"].mean()}, inplace=True)
df.fillna({"Science" : df["Science"].mean()}, inplace=True)
df.fillna({"English" : df["English"].mean()}, inplace=True)

print("After filling missing values:\n", df, "\n")

#Create a column Total = Math + Science + English.
df['Total']=df['Math']+df['English']+df['Science']
print(df)

#Create a column Grade:
#A if Total ≥ 250
#B if Total ≥ 200
#else C
def grade(Total):
    if Total>=250:
        return 'A'
    elif Total>=200:
        return 'B'
    else:
        return 'C'
df['Grade']=df['Total'].apply(grade)
print(df)

# Top score in each city
top_per_city=df.loc[df.groupby('City')['Total'].idxmax()]
print('top scorer in each city:\n', top_per_city,'\n')


#avg score per city per subject
avg_per_city= df.groupby('City')[['Math','Science','English']].max()
print('avg score per subject per city:\n', avg_per_city, '\n')

# students with highest total
high_scorer=df.loc[df['Total'].idxmax()]
print('Top students overall:\n', high_scorer,'\n')


# Pivot table: City vs Subjects
pivot = pd.pivot_table(df, index="City", values=["Math", "Science", "English"], aggfunc="mean")
print("Pivot Table:\n", pivot, "\n")

# Try unstacking
print("Unstacked view:\n", pivot.unstack(), "\n")

# save cleaned data
df.to_csv("students_cleaned.csv", index=False)

# finder function
def find_student(name):
    result = df[df["Name"].str.lower() == name.lower()]
    if result.empty:
        return f"No student found with name {name}"
    return result

print(find_student("Sara"))
```
